## Agrupando os participantes em 5 grupos

**Atribuindo os registros aos devidos grupos:**

```python
kmeans = KMeans(n_clusters=5)
k_fit = kmeans.fit(data)
```

**Inserindo os r√≥tulos dos clusters no dataframe:**

```python
predicoes = k_fit.labels_
data['Clusters'] = predicoes
```

**Verificando os dados**

```python
data.head()
```

## Analisando os grupos

**Qual a quantidade de observa√ß√µes em cada grupo:**

```python
data["Clusters"].value_counts()
```

**Agrupando os registros por grupos**

```python
data.groupby('Clusters').mean()
```

**Calculando a m√©dia de cada grupo de quest√µes para verificar um padr√£o.**

**Selecionando as colunas de cada grupo:**

```python
col_list = list(data)
ext = col_list[0:10]
est = col_list[10:20]
agr = col_list[20:30]
csn = col_list[30:40]
opn = col_list[40:50]
```

**Somando os valores de cada grupo:**

```python
data_soma = pd.DataFrame()
data_soma['extrovertido(a)'] = data[ext].sum(axis=1)/10
data_soma['neur√≥tico(a)'] = data[est].sum(axis=1)/10
data_soma['agrad√°vel(a)'] = data[agr].sum(axis=1)/10
data_soma['meticuloso(a)'] = data[csn].sum(axis=1)/10
data_soma['Abertura para experi√™ncias'] = data[opn].sum(axis=1)/10
data_soma['clusters'] = predicoes
```

**Exibindo o valor m√©dio por grupo:**

```python
data_soma.groupby('clusters').mean()
```

**Visualizando as m√©dias por grupo:**

```python
data_clusters = data_soma.groupby('clusters').mean()
```

```python
plt.figure(figsize=(22,3))
for i in range(0, 5):
    plt.subplot(1,5,i+1)
    plt.bar(data_clusters.columns, data_clusters.iloc[:, i], color='green', alpha=0.2)
    plt.plot(data_clusters.columns, data_clusters.iloc[:, i], color='red')
    plt.title('Grupo ' + str(i))
    plt.xticks(rotation=45)
    plt.ylim(0,4)
```

```python
import gradio as gr
```

**Lendo os dados com as quest√µes.**

```python
dicio_questions = open("questions.txt", encoding="utf-8").read().split("\n")
```

```python
dicio_questions
```

**Limpando dados e recuperando apenas as quest√µes.**

```python
questions = []
for q in dicio_questions:
    q = str(q)
    questions.append(q[q.find("\t"):].lstrip())
```

```python
questions
```

**Criando os inputs dinamicos para passar ao gradio.**

```python
inputs_questions = []
for q in questions:
    obj_input = gr.Slider(minimum=1, maximum=5, step=1, value=3, label=q)
    inputs_questions.append(obj_input)
```

**Verificando os inputs**

```python
inputs_questions
```

**Treinando o modelo para apenas as 50 perguntas**

```python
data = data.drop("Clusters", axis=1)
```

```python
k_fit = kmeans.fit(data)
```

**Criando a interface e a fun√ß√£o predict**

```python
def predict(*outputs_questions):
    outputs_questions = np.array(outputs_questions).reshape(1, -1)
    return k_fit.predict(outputs_questions)

with gr.Blocks() as iface:
    gr.Markdown(
        """
        ## Os Cinco Grandes Tra√ßos de Personalidade  
        Sistema para detec√ß√£o de tra√ßos de personalidade.  
        **LEGENDA BASE DOS GRUPOS:**  
        """
    )
    gr.Image(value="grupos.png", label="Grupos")

    with gr.Row():
        with gr.Column():
            for slider in inputs_questions:
                slider.render()
        with gr.Column():
            output = gr.Textbox(label="Resultado do Modelo")

    gr.Button("Predizer").click(predict, inputs=inputs_questions, outputs=output)
    
iface.launch(share=True)
```
